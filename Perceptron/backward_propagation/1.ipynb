# Diabetes Risk Prediction(Binary Classification)

```python
import numpy as np

np.random.seed(0) 

x=np.array([
    [120,85,150,95],
    [25,30,35,22]
])
```

```python
#Normalize Values
x=x/200

#output : 1-High Risk,0-low Risk
y=np.array([[1,0,1,0]])


print(x.shape)
print(y.shape)
```

# Sigmoid function

```python
def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

# Initialize Parameters

```python
#Hidden Layer
w1=np.random.randn(3,2)*0.1
b1=np.zeros((3,1))

#Output Layer
w2=np.random.randn(1,3)*0.1
b2=np.zeros((1,1))
```

# Forward Propagation


```python
#hidden layer
Z1=np.dot(w1,x)+b1
a1=sigmoid(Z1)

#output layer
Z2=np.dot(w2,a1)+b2
a2=sigmoid(Z2)


print("Output of output layer : ",a2)
```

# Loss Function(Binary Cross Entropy)

```python
m=y.shape[1]
loss=-np.sum(y*np.log(a2)+(1-y)*np.log(1-a2))/m
print("Loss : ",loss)
```

# Back Propagation(output Layer Gradient)

```python
dZ2=a2-y
dw2=(1/m)*np.dot(dZ2,a1.T)
db2=(1/m)*np.sum(dZ2,axis=1,keepdims=True)
```

# Hidden Layer Gradient

```python
da1=np.dot(w2.T,dZ2)
dZ1=da1*a1*(1-a1)

dw1=(1/m)*np.dot(dZ1,x.T)
db1=(1/m)*np.sum(dZ1,axis=1,keepdims=True)
```

# Update Parameters

```python
learning_rate=0.5
w2=w2-learning_rate*dw2
b2=b2-learning_rate*db2

w1=w1-learning_rate*dw1
b1=b1-learning_rate*db1
```

# Complete Trainning Loop

```python
epochs=5000
for i in range(epochs):
    #forward propogation
    Z1=np.dot(w1,x)+b1
    a1=sigmoid(Z1)

    Z2=np.dot(w2,a1)+b2
    a2=sigmoid(Z2)

    #loss
    m=y.shape[1]
    loss=-np.sum(y*np.log(a2)+(1-y)*np.log(1-a2))/m

    #backward propogation
    dZ2=a2-y
    dw2=(1/m)*np.dot(dZ2,a1.T)
    db2=(1/m)*np.sum(dZ2,axis=1,keepdims=True)

    da1=np.dot(w2.T,dZ2)
    dZ1=da1*a1*(1-a1)

    dw1=(1/m)*np.dot(dZ1,x.T)
    db1=(1/m)*np.sum(dZ1,axis=1,keepdims=True)

    #update weights and bias
    learning_rate=0.5
    w2=w2-learning_rate*dw2
    b2=b2-learning_rate*db2

    w1=w1-learning_rate*dw1
    b1=b1-learning_rate*db1
    if i%500==0:
        print("Epoch : ",i," Loss : ",loss)
        
```
