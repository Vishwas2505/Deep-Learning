```python
import numpy as np
```

```python
X=np.array([
    [1,1],
    [1,0],
    [0,1],
    [0,0]
])
y=np.array([1,1,1,0])

X,y
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10

print("Initial Weight:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
    return 1 if z>=0 else 0
```

```python
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_hat = 1 if z > 0 else 0
        
        # Update rule
        w = w + learning_rate * (y[i] - y_hat) * X[i]
        b = b + learning_rate * (y[i] - y_hat)

print("Final weights:", w)
print("Final bias:", b)
```

```python
# Separate spam and non-spam points
import matplotlib.pyplot as plt
spam = X[y == 1]
not_spam = X[y == 0]

plt.scatter(spam[:,0], spam[:,1], marker='o', label='Spam (1)')
plt.scatter(not_spam[:,0], not_spam[:,1], marker='x', label='Not Spam (0)')

plt.xlabel("Suspicious Keywords (x1)")
plt.ylabel("Unknown Sender (x2)")
plt.legend()
plt.grid(True)
```

```python
# Generate x values
x_values = np.linspace(-0.5, 1.5, 100)

# Calculate decision boundary
y_values = -(w[0]*x_values + b) / w[1]

plt.scatter(spam[:,0], spam[:,1], marker='o', label='Spam (1)')
plt.scatter(not_spam[:,0], not_spam[:,1], marker='x', label='Not Spam (0)')
plt.plot(x_values, y_values, label='Decision Boundary')

plt.xlabel("Suspicious Keywords (x1)")
plt.ylabel("Unknown Sender (x2)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_modified = np.array([1, 1, 0, 0])
```

```python
w2 = np.zeros(2)
b2 = 0

for epoch in range(20):
    for i in range(len(X)):
        z = np.dot(w2, X[i]) + b2
        y_hat = 1 if z > 0 else 0
        
        w2 = w2 + learning_rate * (y_modified[i] - y_hat) * X[i]
        b2 = b2 + learning_rate * (y_modified[i] - y_hat)

print("Weights after modified training:", w2)
print("Bias after modified training:", b2)
```
