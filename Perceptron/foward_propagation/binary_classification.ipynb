# Binary Classification
1->pass
0->fail

```python
import numpy as np
```

# Step 1:Prepare Trainig Data

```python
#Traning Data
x=np.array([1,2,3,4])  #StudyHours
y=np.array([0,0,1,1])  
```

# Step 2:Initialize Parameters

```python
w=0.0
b=0.0
learning_rate=1
epochs=5

print("Initial Weight:",w)
print("Initial Bias:",b)
```

# DefineActivation Function(Step Function)

```python
def step_function(z):
    return 1 if z>=0 else 0
```

# Traiing Loop

```python
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    for i in range(len(x)):
        #compute the Weighted Sum
        z=w*x[i]+b
        
        #Apply the step function to get the predicted output
        y_pred=step_function(z)
        
        #Calculate the error
        error=y[i]-y_pred
        
        #Update the weights and bias
        w=w+learning_rate*error*x[i]
        b=b+learning_rate*error
        
        print(f"Input: {x[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}, Updated Weight: {w}, Updated Bias: {b}")

```

# final model parameters

```python
print("Final Weight:",w)
print("Final Bias:",b)
```

# Testing value

```python
test_value=3.1
z=w*test_value+b
y_pred=step_function(z)
print(f"Test Input: {test_value}, Predicted Output: {y_pred}")
```

```python
test_value=2.9
z=w*test_value+b
y_pred=step_function(z)
print(f"Test Input: {test_value}, Predicted Output: {y_pred}")
```

```python
#plot with decision boundary
import matplotlib.pyplot as plt
plt.scatter(x,y,color='blue',label='Data Points')
x_values=np.array([0,5])
y_values=-(w*x_values+b)/w
plt.plot(x_values,y_values,color='red',label='Decision Boundary')
plt.xlabel('Study Hours')
plt.ylabel('Pass (1) / Fail (0)')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid()
plt.show()
```
