```python
import numpy as np
```

```python
x=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y=np.array([0,1,1,1])
x,y
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10

print("Initial Weights:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
    return 1 if z>=0 else 0
```

```python
for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")
    for i in range(len(x)):
        linear_output=np.dot(w,x[i])+b
        y_pred=step_function(linear_output)

        update=learning_rate*(y[i]-y_pred)
        w+=update*x[i]
        b+=update
        print(f"Sample {i+1}, Weights: {w}, Bias: {b}")
```

```python
print("Final Weight:",w)
print("Final Bias:",b)
```

```python
test_values=[[0,0],[0,1],[1,0],[1,1]]
for i in range(len(x)):
    z=np.dot(w,x[i])+b
    y_pred=step_function(z)
    print(f"Input: {x[i]}, Predicted Output: {y_pred}, Actual Output: {y[i]}")
```

```python
#plotting decision boundary
import matplotlib.pyplot as plt
x1=np.array([0,1])
x2=-(w[0]*x1+b)/w[1]
plt.plot(x1,x2,label="Decision Boundary")
plt.scatter(x[:,0],x[:,1],c=y,label="Data Points")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.title("OR Gate Decision Boundary")
plt.legend()
plt.grid()
plt.show()
```
