```python
import numpy as np 
```

```python
x=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,1,1,0])

x,y
```

```python
#Decalaration 
w1=1.0
w2=1.0
w3=1.0
w4=1.0
w5=1.0
w6=-2.0
b1=-0.5
b2=-1.5
b3=-0.5
def stepfunction(x):
    if x>=0:
        return 1
    else:
        return 0
    
```

```python
def forward(x):
    h1=(w1*x[0]+w2*x[1]+b1)
    h2=(w3*x[0]+w4*x[1]+b2)
    y=(w5*stepfunction(h1)+w6*stepfunction(h2)+b3)
    return y
for i in range(4):
    print(stepfunction(forward(x[i])))
```

```python
#plot decision boundary
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('XOR')
plt.show()
```

```python
#test_value
print(stepfunction(forward(np.array([0,0]))))
print(stepfunction(forward(np.array([0,1]))))
print(stepfunction(forward(np.array([1,0]))))
print(stepfunction(forward(np.array([1,1]))))
```
